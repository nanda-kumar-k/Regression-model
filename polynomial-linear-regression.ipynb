```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data_set = pd.read_csv('data/Position_Salaries.csv')
x = data_set.iloc[:, 1:-1].values
y = data_set.iloc[:,-1].values
```

```python
# simple Linear Regression model on whole data set 
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```python
# Polynomial Linear Regression
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4)
x_ploy = poly_reg.fit_transform(x)

print(x_ploy)
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_ploy, y)
```

```python
plt.scatter(x,y, color='red')
plt.plot(x, lin_reg.predict(x), color='blue')
plt.title('Truth or Bluff LinearRegression')
plt.xlabel('Postiton Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y, color='red')
plt.plot(x, lin_reg_2.predict(x_ploy), color='blue')
plt.title('Truth or Bluff Polynomial Regression')
plt.xlabel('Postiton Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression results (for higher resolution and smoother curve)

X_grid = np.arange(min(x), max(x), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# predicting the value for linear regression model
lin_reg.predict([[6.5]])
```

```python
# predicting the value for polynomial linear regression model
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
